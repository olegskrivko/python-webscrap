```python
driver = webdriver.Firefox(executable_path=r"C:\Users\olegs\Documents\geckodriver-v0.31.0-win64\geckodriver")
driver.get("https://www.ss.lv/lv/")

#implicit wait
driver.implicitly_wait(0.5)
#maximize browser
driver.maximize_window()

#btn = driver.find_element(By.ID, "mtd_59")
btn = driver.find_element(By.XPATH,'//a[contains(@href,"/lv/real-estate/flats/")]')
btn.click()
random_time = (random.random() + random.randint(1,4))
print("Dzīvokļi")


btn = driver.find_element(By.ID, "ahc_14195")
btn.click()
random_time = (random.random() + random.randint(1,4))
print("Rīga")

btn = driver.find_element(By.ID, "ahc_1106")
btn.click()
random_time = (random.random() + random.randint(1,4))
print("Centrs")

#filter price
#filter_price_min = driver.find_element(By.ID, "f_o_8_min")
filter_price_min = driver.find_element(By.XPATH,f'//input[contains(@name,"topt[8][min]")]')
#clear old value if it was unclear
#filter_price_min.click();

filter_price_min.clear()
#enter min price
filter_price_min.send_keys(price_min)
#see letters in input
filter_price_min.send_keys(Keys.RETURN)
random_time = (random.random() + random.randint(1,4))
time.sleep(random_time)
print(f"Min price entered {price_min}")

#filter price
filter_price_max = driver.find_element(By.ID, "f_o_8_max")
#clear old value if it was unclear
#filter_price_max.click();

filter_price_max.clear()
#enter max price
filter_price_max.send_keys(price_max)
#see letters in input
filter_price_max.send_keys(Keys.RETURN)
random_time = (random.random() + random.randint(1,4))
time.sleep(random_time)
print(f"Max price entered {price_max}")


#filter rooms
sel = Select(driver.find_element(By.CSS_SELECTOR, 'select.l50:nth-child(1)'))
#select by select_by_visible_text() method
sel.select_by_visible_text(rooms_min)
random_time = (random.random() + random.randint(1,4))
print(f"Min rooms chosen {rooms_min}")

#filter rooms
#sel = Select(driver.find_element(By.XPATH, '//*[@id="f_o_1"]'))
sel = Select(driver.find_element(By.CSS_SELECTOR, 'select.l50:nth-child(2)'))
#select by select_by_visible_text() method
sel.select_by_visible_text(rooms_max)
random_time = (random.random() + random.randint(1,4))
time.sleep(random_time)
print(f"Max rooms chosen {rooms_max}")


#filter area
filter_area_min = driver.find_element(By.ID, "f_o_3_min")
#clear old value if it was unclear
filter_area_min.click()

filter_area_min.clear()
#enter min price
filter_area_min.send_keys(area_min)
#see letters in input
filter_area_min.send_keys(Keys.RETURN)
random_time = (random.random() + random.randint(1,4))
time.sleep(random_time)
print(f"Min area entered {area_min}")

#filter area
filter_area_max = driver.find_element(By.ID, "f_o_3_max")
#clear old value if it was unclear
filter_area_max.click()

filter_area_max.clear()
#enter max price
filter_area_max.send_keys(area_max)
#see letters in input
filter_area_max.send_keys(Keys.RETURN)
random_time = (random.random() + random.randint(1,4))
time.sleep(random_time)
print(f"Max area entered {area_max}")

#filter floor
filter_floor_min = driver.find_element(By.ID, "f_o_4_min")
#clear old value if it was unclear
filter_floor_min.click()

filter_floor_min.clear()
#enter min price
filter_floor_min.send_keys(floor_min)
#see letters in input
filter_floor_min.send_keys(Keys.RETURN)
random_time = (random.random() + random.randint(1,4))
time.sleep(random_time)
print(f"Min floor entered {floor_min}")

#filter floor
filter_area_max = driver.find_element(By.ID, "f_o_4_max")
#clear old value if it was unclear
filter_area_max.click()

filter_area_max.clear()
#enter max price
filter_area_max.send_keys(floor_max)
#see letters in input
filter_area_max.send_keys(Keys.RETURN)
random_time = (random.random() + random.randint(1,4))
time.sleep(random_time)
print(f"Max floor entered {floor_max}")


#building type
sel = Select(driver.find_element(By.ID, 'f_o_6'))
#select by select_by_visible_text() method
sel.select_by_visible_text(building_type[-1])
random_time = (random.random() + random.randint(1,4))
time.sleep(random_time)
print(f"Building type chosen {building_type[-1]}")


#filter deal type
#sel = Select(driver.find_element(By.XPATH, '/html/body/div[4]/div/table/tbody/tr/td/div[1]/table/tbody/tr/td/form/div/span[2]/select'))
sel = Select(driver.find_element(By.CLASS_NAME, 'l100'))
#select by select_by_visible_text() method
sel.select_by_visible_text(deal_type[3])
random_time = (random.random() + random.randint(1,4))
time.sleep(random_time)
print(f"Deal type chosen {deal_type[3]}")


#filter district
sel = Select(driver.find_element(By.XPATH, '/html/body/div[4]/div/table/tbody/tr/td/div[1]/table/tbody/tr/td/form/div/span[3]/select'))
#select by select_by_visible_text() method
sel.select_by_visible_text(disctrict[0])
random_time = (random.random() + random.randint(1,4))
time.sleep(random_time)
print(f"District chosen {disctrict[0]}")




# #button
# submit_btn = driver.find_element(By.XPATH, "/html/body/div[4]/div/table/tbody/tr/td/div[1]/table/tbody/tr/td/form/table[1]/tbody/tr/td[2]/input")
# submit_btn.click();

# random_time = (random.random() + random.randint(1,4))
# time.sleep(random_time)
# print("Submit btn clicked")

#row_element_count
row_element = driver.find_elements(By.XPATH, "/html/body/div[4]/div/table/tbody/tr/td/div[1]/table/tbody/tr/td/form/table[2]/tbody/tr")
row_element_count = len(row_element)

#print(f"Found {row_element_count} items")

#row_element_count
td_element = driver.find_elements(By.XPATH, f"/html/body/div[4]/div/table/tbody/tr/td/div[1]/table/tbody/tr/td/form/table[2]/tbody/tr[2]/td")
td_element_count = len(td_element)

#print(f"Found {td_element_count} td items")
total_results = []

for x in range(2,row_element_count):
    item_row = driver.find_element(By.XPATH, f"/html/body/div[4]/div/table/tbody/tr/td/div[1]/table/tbody/tr/td/form/table[2]/tbody/tr[{x}]")
    item_row_data = item_row.get_attribute("innerHTML")
    
    flat_list = []
    for y in range(4,(td_element_count + 1)):
        tr_td_element = driver.find_element(By.XPATH, f"/html/body/div[4]/div/table/tbody/tr/td/div[1]/table/tbody/tr/td/form/table[2]/tbody/tr[{x}]/td[{y}]/a")
        tr_td_element_data = tr_td_element.get_attribute("innerHTML")
        
        #print(tr_td_element_link)
        flat_list.append(tr_td_element_data)
    tr_td_element_link = tr_td_element.get_attribute('href')
    #print(tr_td_element_link)
    flat_list.append(tr_td_element_link)
    total_results.append(flat_list)
    
header = ['Iela', 'Istabas', 'Platiba', 'Stavs', 'Serija', 'Cena,m2', 'Cena']
print(header)    
for item in total_results:
    print(item)

driver.close()
print('Script ended')
# select by visible text
# select.select_by_visible_text('Banana')

# select by value 
# select.select_by_value('1')
```
